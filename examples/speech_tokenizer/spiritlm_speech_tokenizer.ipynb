## Load audio

```python
import IPython.display as ipd
audio = "../audio/7143-88743-0029.flac"
print('Original audio:')
ipd.display(ipd.Audio(audio))
```

## Spiritlm tokenizers

```python
from spiritlm.speech_tokenizer import spiritlm_base, spiritlm_expressive
```

### SpiritLM-BASE tokenizer

```python
## Load the tokenizer
tokenizer_base = spiritlm_base()
```

```python
## encode_units
print('SpiritLM-BASE: Encode audio into units (not deduplicated) \n', '-'*20)
units = tokenizer_base.encode_units(audio)
print(units)

## encode_string
print('\nSpiritLM-BASE: Encode audio into string (deduplicated and sorted units) \n', '-'*20)
string_tokens = tokenizer_base.encode_string(audio)
print(string_tokens)

## decode from units
print('\nSpiritLM-BASE: Decode back to audio from units (not deduplicated) \n', '-'*20)
resyn_wav = tokenizer_base.decode(units, speaker_id=2)
ipd.display(ipd.Audio(resyn_wav, rate=16000))

## decode from string
print('\nSpiritLM-BASE: Decode back to audio from string (deduplicated and sorted units) \n', '-'*20)
resyn_dedup_wav = tokenizer_base.decode(string_tokens, speaker_id=2)
ipd.display(ipd.Audio(resyn_dedup_wav, rate=16000))
```

### SpiritLM-EXPRESSIVE Tokenizer

```python
## Load the tokenizer
tokenizer_expressive = spiritlm_expressive()
```

```python
## encode_units
print('SpiritLM-EXPRESSIVE: Encode audio into units (not deduplicated) \n', '-'*20)
units = tokenizer_expressive.encode_units(audio)
print(units)

## encode_string
print('\nSpiritLM-EXPRESSIVE: Encode audio into string (deduplicated and sorted units) \n', '-'*20)
string_tokens = tokenizer_expressive.encode_string(audio)
print(string_tokens)

## decode from units
print('\nSpiritLM-EXPRESSIVE: Decode back to audio from units (not deduplicated) \n', '-'*20)
resyn_wav = tokenizer_expressive.decode(units, speaker_id=2)
ipd.display(ipd.Audio(resyn_wav, rate=16000))

## decode from string
print('\nSpiritLM-EXPRESSIVE: Decode back to audio from string (deduplicated and sorted units) \n', '-'*20)
resyn_dedup_wav = tokenizer_expressive.decode(string_tokens, speaker_id=2)
ipd.display(ipd.Audio(resyn_dedup_wav, rate=16000))
```

```python
from spiritlm.speech_tokenizer.hubert import spirit_hubert
from spiritlm.speech_tokenizer.hifigan import spirit_base_hifigan, spirit_expressive_hifigan
from spiritlm.speech_tokenizer.f0 import spirit_expressive_f0
from spiritlm.speech_tokenizer.style_encoder import spirit_expressive_w2v2
```

```python
hubert = spirit_hubert()
pitcher = spirit_expressive_f0()
styler = spirit_expressive_w2v2()
voc = spirit_base_hifigan()
voc_ex = spirit_expressive_hifigan()
```

```python
hubert(audio)
```

```python
wav = voc(hubert(audio), speaker_id=1).cpu().numpy()
ipd.Audio(wav, rate=16000)
```

```python
styler(audio)
```

```python
wav_ex = voc_ex(
            code=hubert(audio),
            f0_code=pitcher(audio),
            style_code=styler(audio),
            dur_pred=False,
            speaker_id=1,
            not_dedup_code=True,
        ).cpu().numpy()
ipd.Audio(wav_ex, rate=16000)
```
